```python
### Open AI API Key and Open Source models--Llama3,Gemma2,mistral--Groq

import os
from dotenv import load_dotenv
load_dotenv()

import openai
openai.api_key=os.getenv("OPENAI_API_KEY")

groq_api_key=os.getenv("GROQ_API_KEY")
# groq_api_key
```

```python
%pip install langchain_groq
```

```python
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
model=ChatGroq(model="llama-3.3-70b-versatile",groq_api_key=groq_api_key)
# model =ChatOpenAI(model="gpt-3.5-turbo",temperature=0.9)
model
```

```python
%pip install langchain_core
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

#HumanMessage and SystemMessage. The SystemMessage is used to provide instructions or context to the model, while the HumanMessage represents the user's input.

messages=[
    SystemMessage(content="Translate the following from English to Telugu"),
    HumanMessage(content="Had your lunch?")
]

result=model.invoke(messages)
```

```python
result
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(result)
```

```python
### Using LCEL- chain the components
chain=model|parser
chain.invoke(messages)
```

```python
### Prompt Templates
from langchain_core.prompts import ChatPromptTemplate

generic_template="Trnaslate the following into {language}:"

prompt=ChatPromptTemplate.from_messages(
    [("system",generic_template),("user","{text}")]
)

```

```python
result=prompt.invoke({"language":"Telugu","text":"how are you?"})
```

```python
result.to_messages()
# result
```

```python
##Chaining together components with LCEL
chain=prompt|model|parser
chain.invoke({"language":"Telugu","text":"How are you?"})
```
