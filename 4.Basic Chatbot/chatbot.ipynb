### Message History
We can use a Message History class to wrap our model and make it stateful. This will keep track of inputs and outputs of the model, and store them in some datastore. Future interactions will then load those messages and pass them into the chain as part of the input. Let's see how to use this!

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    # Check if this session already has a chat history
    if session_id not in store:

        # If not, create a new empty chat history for this session
        store[session_id] = ChatMessageHistory()

    # Return the existing or newly created chat history
    return store[session_id]

with_message_history=RunnableWithMessageHistory(model,get_session_history)
```

```python
from IPython.display import display, HTML
from html import escape

from langchain_core.messages import (
    HumanMessage,
    AIMessage,
    SystemMessage,
    ToolMessage,
)


# Get the chat history
testing = get_session_history("chat1")


def display_chat_history(chat_history):
    """
    Display LangChain chat history in a clean chat-style layout.

    Parameters
    ----------
    chat_history:
        An InMemoryChatMessageHistory or ChatMessageHistory object.
    """

    if not chat_history.messages:
        display(
            HTML(
                """
                <div style="
                    padding: 16px;
                    border: 1px solid #d1d5db;
                    border-radius: 12px;
                    font-family: Arial, sans-serif;
                ">
                    No messages are available in this chat session.
                </div>
                """
            )
        )
        return

    formatted_messages = []

    for index, message in enumerate(chat_history.messages, start=1):

        # Identify the message type
        if isinstance(message, HumanMessage):
            role = "Human"
            icon = "👤"
            role_class = "human-message"

        elif isinstance(message, AIMessage):
            role = "AI"
            icon = "🤖"
            role_class = "ai-message"

        elif isinstance(message, SystemMessage):
            role = "System"
            icon = "⚙️"
            role_class = "system-message"

        elif isinstance(message, ToolMessage):
            role = "Tool"
            icon = "🛠️"
            role_class = "tool-message"

        else:
            role = message.__class__.__name__
            icon = "💬"
            role_class = "other-message"

        # Escape HTML characters for safe rendering
        content = escape(str(message.content))

        # Preserve line breaks
        content = content.replace("\n", "<br>")

        # Get token information when available
        usage_metadata = getattr(message, "usage_metadata", None) or {}

        input_tokens = usage_metadata.get("input_tokens")
        output_tokens = usage_metadata.get("output_tokens")
        total_tokens = usage_metadata.get("total_tokens")

        token_details = ""

        if total_tokens is not None:
            token_details = f"""
                <div class="token-info">
                    <span>Input tokens: {input_tokens or 0}</span>
                    <span>Output tokens: {output_tokens or 0}</span>
                    <span>Total tokens: {total_tokens}</span>
                </div>
            """

        formatted_messages.append(
            f"""
            <div class="message-row {role_class}">
                <div class="message-card">

                    <div class="message-header">
                        <div class="role-details">
                            <span class="role-icon">{icon}</span>
                            <span class="role-name">{role}</span>
                        </div>

                        <span class="message-number">
                            Message {index}
                        </span>
                    </div>

                    <div class="message-content">
                        {content}
                    </div>

                    {token_details}

                </div>
            </div>
            """
        )

    complete_html = f"""
    <style>
        .chat-history-container {{
            max-width: 950px;
            margin: 20px auto;
            padding: 24px;
            border: 1px solid color-mix(
                in srgb,
                currentColor 16%,
                transparent
            );
            border-radius: 18px;
            background: color-mix(
                in srgb,
                Canvas 97%,
                currentColor 3%
            );
            color: CanvasText;
            font-family:
                Inter,
                -apple-system,
                BlinkMacSystemFont,
                "Segoe UI",
                Roboto,
                Arial,
                sans-serif;
        }}

        .chat-history-title {{
            margin: 0;
            font-size: 26px;
            font-weight: 800;
            letter-spacing: -0.4px;
        }}

        .chat-history-subtitle {{
            margin: 6px 0 24px;
            font-size: 14px;
            opacity: 0.7;
        }}

        .message-row {{
            display: flex;
            margin-bottom: 16px;
        }}

        .human-message {{
            justify-content: flex-end;
        }}

        .ai-message,
        .system-message,
        .tool-message,
        .other-message {{
            justify-content: flex-start;
        }}

        .message-card {{
            width: min(76%, 720px);
            padding: 16px 18px;
            border: 1px solid color-mix(
                in srgb,
                currentColor 14%,
                transparent
            );
            border-radius: 16px;
            box-shadow: 0 5px 16px rgba(0, 0, 0, 0.06);
        }}

        .human-message .message-card {{
            background: color-mix(
                in srgb,
                Canvas 84%,
                #2563eb 16%
            );
            border-bottom-right-radius: 5px;
        }}

        .ai-message .message-card {{
            background: color-mix(
                in srgb,
                Canvas 86%,
                #10b981 14%
            );
            border-bottom-left-radius: 5px;
        }}

        .system-message .message-card {{
            background: color-mix(
                in srgb,
                Canvas 86%,
                #8b5cf6 14%
            );
            border-left: 4px solid #8b5cf6;
        }}

        .tool-message .message-card {{
            background: color-mix(
                in srgb,
                Canvas 86%,
                #f59e0b 14%
            );
            border-left: 4px solid #f59e0b;
        }}

        .message-header {{
            display: flex;
            align-items: center;
            justify-content: space-between;
            gap: 15px;
            margin-bottom: 10px;
        }}

        .role-details {{
            display: flex;
            align-items: center;
            gap: 8px;
        }}

        .role-icon {{
            font-size: 18px;
        }}

        .role-name {{
            font-size: 14px;
            font-weight: 800;
        }}

        .message-number {{
            font-size: 11px;
            font-weight: 700;
            opacity: 0.55;
        }}

        .message-content {{
            font-size: 15px;
            line-height: 1.7;
            overflow-wrap: anywhere;
        }}

        .token-info {{
            display: flex;
            flex-wrap: wrap;
            gap: 8px;
            margin-top: 13px;
            padding-top: 11px;
            border-top: 1px solid color-mix(
                in srgb,
                currentColor 12%,
                transparent
            );
        }}

        .token-info span {{
            padding: 4px 8px;
            border-radius: 7px;
            background: color-mix(
                in srgb,
                Canvas 86%,
                currentColor 14%
            );
            font-size: 10px;
            font-weight: 700;
            opacity: 0.76;
        }}

        @media (max-width: 700px) {{
            .chat-history-container {{
                padding: 16px;
            }}

            .message-card {{
                width: 90%;
            }}
        }}

        @media (prefers-color-scheme: dark) {{
            .message-card {{
                box-shadow: 0 5px 18px rgba(0, 0, 0, 0.3);
            }}
        }}
    </style>

    <div class="chat-history-container">

        <h2 class="chat-history-title">
            Chat Session: chat1
        </h2>

        <p class="chat-history-subtitle">
            Total messages: {len(chat_history.messages)}
        </p>

        {''.join(formatted_messages)}

    </div>
    """

    display(HTML(complete_html))


# Display the formatted chat history
display_chat_history(testing)
```

### Managing the Conversation History
One important concept to understand when building chatbots is how to manage conversation history. If left unmanaged, the list of messages will grow unbounded and potentially overflow the context window of the LLM. Therefore, it is important to add a step that limits the size of the messages you are passing in.
'trim_messages' helper to reduce how many messages we're sending to the model. The trimmer allows us to specify how many tokens we want to keep, along with other parameters like if we want to always keep the system message and whether to allow partial messages

```python
from langchain_core.messages import SystemMessage,trim_messages
trimmer=trim_messages(
    max_tokens=45,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human"
)
messages = [
    SystemMessage(content="you're a good assistant"),

    HumanMessage(content="hi! I'm bob"),
    AIMessage(content="hi!"),
    HumanMessage(content="I like vanilla ice cream"),
    AIMessage(content="nice"),
    HumanMessage(content="whats 2 + 2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="no problem!"),
    HumanMessage(content="having fun?"),
    AIMessage(content="yes!"),
]
trimmer.invoke(messages)
```

```python
from operator import itemgetter

from langchain_core.runnables import RunnablePassthrough

chain=(
    RunnablePassthrough.assign(messages=itemgetter("messages")|trimmer)
    | prompt
    | model
    
)

response=chain.invoke(
    {
    "messages":messages + [HumanMessage(content="What ice cream do i like")],
    "language":"English"
    }
)
response.content
```
